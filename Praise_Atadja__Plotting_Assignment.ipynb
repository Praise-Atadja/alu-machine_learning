```python
# Plot a histogram for `co2` - Carbon Dioxide values
column_values = fuel_econ['co2']
plt.figure(figsize=(10, 6))
plt.hist(column_values, bins=50, edgecolor='black', color='pink')
plt.title('Distribution of Carbon Dioxide (co2) Values')
plt.xlabel('Co2 Values')
plt.ylabel('Measure')
plt.show()

# Check skewness for the `co2` distribution and define if it's skewed or symmetric
skewness = fuel_econ['co2'].skew()
if abs(skewness) > 0.5:
    skew_type = "skewed"
else:
    skew_type = "symmetric"

print(f"The distribution is {skew_type}.")


```

```python
# Plot a histogram for `city` - City Distribution
column_values = fuel_econ['city']
plt.figure(figsize=(10, 6))
plt.hist(column_values, bins=50, edgecolor='black', color='green')
plt.title('Distribution Over City')
plt.xlabel('City Values')
plt.ylabel('Measure')
plt.show()

# Check skewness for the `city` distribution and define if it's skewed or symmetric
skewness = fuel_econ['city'].skew()
if abs(skewness) > 0.5:
    skew_type = "skewed"
else:
    skew_type = "symmetric"

print(f"The distribution is {skew_type}.")
```

**TO DO:** create a heat map of engine displacement i.e the column "disp"  vs. co2 production.and interpret it (is there correlation?what did you notice?). Engine Displacement is in litres

```python
# Create a heatmap
plt.figure(figsize=(8, 6))
sb.heatmap(data=fuel_econ[['displ', 'co2']].corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Co2 Production (g/mi)')
plt.title('Heatmap of Engine Displacement vs. Co2 Production')
plt.show()
```
